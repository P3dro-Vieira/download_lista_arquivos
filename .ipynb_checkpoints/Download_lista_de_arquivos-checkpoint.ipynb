```python
#IMPORTS
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from datetime import datetime, timedelta
import time


# ABRINDO O NAVEGADOR

driver = webdriver.Chrome()
driver.maximize_window()
driver.get('url')
time.sleep(1)


# SELECIONANDO A DATA D-1

# 1. Data de ontem
y = datetime.today() - timedelta(days = 1)


# 2. Dia da semana que foi ontem
dia_semana = y.weekday()


# 3. Avaliar se o dia da semana está entre 0 e 4. Final de semana é 5 e 6.
if dia_semana >= 0 and dia_semana <= 4:
    
    d = y.strftime('%d')
    m = y.strftime('%m')
    a = y.strftime('%Y')
    
    data_alvo = f'{d}/{m}/{a}'
    print(data_alvo)
    
elif dia_semana == 6:
    
    y = datetime.today() - timedelta(days = 3)
    
    d = y.strftime('%d')
    m = y.strftime('%m')
    a = y.strftime('%Y')
    
    data_alvo = f'{d}/{m}/{a}'
        
else:    
    
    time.sleep(3)
    driver.quit()


# LOGIN

username = "seu_usuario"
password = "sua_senha"

driver.find_element(By.ID, 'user').send_keys(username)
time.sleep(1)
driver.find_element(By.ID, 'password').send_keys(password)
time.sleep(1)
driver.find_element(By.CLASS_NAME, 'botao_login').click()


# LISTA DOS FUNDOS A SEREM BAIXADOS

fundos_1 = ['fundo_1', 'fundo_2', 'fundo_3', 'fundo_4','fundo_5', 'fundo_6']


# CREDENCIAIS

confirmacao = driver.find_element(By.CLASS_NAME, 'confirmacao')
login = False

while login == False:
    
    if "Condicao_1" in confirmacao.text:
        
        time.sleep(1)
        driver.find_element(By.NAME, 'cancelar').click()
        time.sleep(1)
        # Volta p/ pagina anterior e digita novamente usuario e senha
        driver.find_element(By.ID, 'id').send_keys(username)
        time.sleep(1)
        driver.find_element(By.ID, 'id').send_keys(password)
        time.sleep(1)
        driver.find_element(By.CLASS_NAME, 'botao_login').click()
        time.sleep(1)
        confirmacao = driver.find_element(By.CLASS_NAME, 'class_name')
        # Esse novo 'confirmacao' é necessário para atualizar o valor da variável. 
        # Assim, ela pode percorrer os outros IFs.
 
    elif "Condicao_2" in confirmacao.text:

        time.sleep(1)
        driver.find_element(By.NAME, 'cancelar').click()
        time.sleep(1)
        # Volta p/ pagina anterior e digita novamente usuario e senha
        driver.find_element(By.ID, 'id').send_keys(username)
        time.sleep(1)
        driver.find_element(By.ID, 'id').send_keys(password)
        time.sleep(1)
        driver.find_element(By.CLASS_NAME, 'notao_login').click()
        time.sleep(1)
        confirmacao = driver.find_element(By.CLASS_NAME, 'class_name')

    else:
        
        login = True
        time.sleep(1)
        # Preenche as credencias e confirma clicando no botão e efetuando o login
        driver.find_element(By.ID, 'sobrenome').send_keys('seu_sobrenome')
        time.sleep(1)
        driver.find_element(By.ID, 'id').click()

        # Selecinando cada fundo e gerando primeiro o arquivo para download
        
        for fundo in fundos_1:

            # Clica em na aba que estão os fundos
            driver.find_element(By.XPATH, 'xpath').click()
            time.sleep(1)

            # Clica na barra de seleçao dos fundos
            driver.find_element(By.ID, 'id').click()
            time.sleep(1)

            # Seleciona um fundo da lista
            driver.find_element(By.ID, fundo).click()
            time.sleep(1)

            # Envia a data já definida acima (data_alvo)
            driver.find_element(By.NAME, 'name').click()
            driver.find_element(By.NAME, 'name').send_keys(data_alvo, Keys.ENTER)
            time.sleep(1)
            driver.find_element(By.ID, 'id').click()
            time.sleep(1)
         
        
        # Clica na aba onde os arquivos foram gerados
        
        driver.find_element(By.XPATH, 'xpath').click()
        time.sleep(60)
        driver.find_element(By.ID, 'id').click()
        time.sleep(2)

            
        # Baixando todos os arquivos
                
        funds = driver.find_elements(By.CLASS_NAME, 'class_name')
        print(len(funds)) # Opcional
        
        # O count foi usado pq sempre vou querer baixar os 6 ultimos fundos gerados.
        # Caso um fundo não tenha sido gerado, ele é pulado, mas mesmo assim é contabilizado,
        # para respeitar essa regra
        count = 0
                
        for fundo in funds:
            
            try:
                
                if count < 6:
                    
                    time.sleep(0.5)
                    # Se o arquivo foi gerado, clica para baixar
                    fundo.find_element(By.ID, 'id').click() 
                    count += 1
                    
                else:
                    
                    break
                    # O BREAK sai do loop
                    
            except:
                # Se o arquivo não foi gerado, contabiliza mesmo assim o count
                count += 1
                
                if count == 6:
                    print("Encerrando...") # Opcional
                    break

                    
time.sleep(15)
driver.close()
```
